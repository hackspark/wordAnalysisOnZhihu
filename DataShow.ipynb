```scala
case class MatchData(id:String, ts:String, words:Array[String])
def Parse(line:String) = {
    val id = line.substring(0, 34)
    var ts = line.substring(34, 44)
    var words = line.substring(44).split(" ")
    MatchData(id, ts, words)
}
```

```scala
val parsed = rawFile.map(line => Parse(line))
```

```scala
val textFile = sc.textFile("matchData").map(_.split("\n"))
val counts = textFile.filter(_(0).length>44).flatMap(line => line(0).substring(44).split(" ")).map(word => (word, 1)).reduceByKey(_ + _)
val top = counts.collect.toSeq.sortBy(_._2).reverse.take(1000)
```

```scala
top.foreach(println)
```
