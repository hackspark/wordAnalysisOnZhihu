```python
import numpy as np
import pandas as pd
import jieba as jb
```

```python
username = "zord-vczh"
f = open(username + "_answerInfos.csv")
infoTexts = f.readlines()
f.close()
f = open(username + "_answerDetails.csv")
detailTexts = f.read().split("]>]>]>")
f.close()
```

```python
slLinfo = []
aL = []
tL = []
for i in range(0, len(infoTexts)):
    tmp = infoTexts[i][:-1].split(",")
    if len(tmp) == 4:
        slLinfo.append(tmp[1])
        aL.append(tmp[2])
        tL.append(tmp[3])
```

```python
slL = []
cL = []
spL = []
for i in range(0, len(detailTexts)):
    tmp = detailTexts[i].split(",|", 2)
    if len(tmp) == 3:
        slL.append(tmp[1])
        cL.append(tmp[2])
        res = jb.cut(tmp[2], cut_all=False)
        res = " ".join(res)
        spL.append(res)
```

```python
totalDict = {
    'shortLink' : pd.Series(slL),
    'content'   : pd.Series(cL),
    'approveNum': pd.Series(aL),
    'time'      : pd.Series(tL),
    'splitRes'  : pd.Series(spL)
    }
```

```python
import pickle
f = open(username + '_dict.data', 'w')
pickle.dump(totalDict, f)
f.close()

#f = open('totalDict', 'r')
#Dict = pickle.load(f)
#f.close()
```
