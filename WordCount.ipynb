```python
import pickle, re
```

```python
username = "zord-vczh"
f = open(username + '_dict.data', 'r')
rdata = pickle.load(f)
f.close()
```

```python
replaceHTMLTagPattern = re.compile(r'(<((.|\n)*?)>)')
replaceSymbolPattern = re.compile(r'("|\(|\)|\;|&|/|\.|\}|\{|\,|\。|\*|\-|\=|\，|\、|\%)')
replaceLinePattern = re.compile(r'(\n)')
replaceSpacePattern = re.compile(r'( )+')
#res = re.sub(replaceHTMLTagPattern, "", s)
```

```python
copydata = {
    'shortLink' : rdata['shortLink'],
    'splitRes'  : rdata['splitRes'],
    'time'      : rdata['time']
    }
f = open("matchData", 'w')
for i in range(0, len(copydata['time'])):
    item = copydata['time'][i]
    copydata['time'][i] = item.split(" ")[0]
    item = copydata['splitRes'][i]
    res = re.sub(replaceHTMLTagPattern, "", item)
    res = re.sub(replaceLinePattern, "", res)
    copydata['splitRes'][i] = res
    f.write(copydata['shortLink'][i]+copydata['time'][i]+res+'\n')
f.close()
```
